```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt

fig_num = 43
infile=[num for num in range(1,fig_num+2,1)]
image=list(range(1,fig_num+2))

filename="neuron"

for i in range(1, fig_num+1):  
    infile[i] = filename +  str(i+42) +".jpg"

imgname="result-auto-10"
plt.figure(figsize=(30, 40))
plt.subplots_adjust(wspace=0.0, hspace=0.0)

for i in range(1, fig_num+1):  
    image[i] = cv2.imread(infile[i])
    image[i] = cv2.cvtColor(image[i], cv2.COLOR_BGR2RGB)
    plt.subplot(9,5,i)
    plt.subplots_adjust(wspace=0.0, hspace=0.0)
    plt.tick_params(labelbottom="off")
    plt.tick_params(labelleft="off")
    plt.tick_params(bottom="off")
    plt.tick_params(left="off")
    plt.gca().spines['right'].set_visible(False) #
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    #plt.rcParams["font.size"] = 0.1
    plt.xticks(color="None")
    plt.yticks(color="None")
    plt.tick_params(length=0)

    #plt.title( str(i))
    plt.imshow(image[i]) 
plt.savefig(imgname + '.jpg')
```
